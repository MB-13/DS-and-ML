```python
### Feature selection using exhaustive feature seletor method for regression dataset
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,:-1],df['medv'],test_size=0.2,random_state=1)
```

```python
x_train
```

```python
columns = x_train.columns
columns
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train = pd.DataFrame(x_train,columns=columns)
x_test = pd.DataFrame(x_test,columns=columns)
```

```python
x_train
```

```python
x_test
```

```python
#baseline model
import numpy as np
from sklearn.metrics import r2_score
model = LinearRegression()

print("training",np.mean(cross_val_score(model,x_train,y_train,cv=5,scoring='r2')))
print("testing",np.mean(cross_val_score(model,x_test,y_test,cv=5,scoring='r2')))
```

```python
lr = LinearRegression()
exh = EFS(lr,max_features=13,scoring='r2',cv=10,print_progress=True,n_jobs=-1)
sel = exh.fit(x_train,y_train)
```

```python
sel.best_score_
```

```python
columns = sel.best_feature_names_
```

```python
x_train_sel = sel.transform(x_train)
x_train_sel = pd.DataFrame(x_train_sel,columns=columns)
x_test_sel = sel.transform(x_test)
x_test_sel = pd.DataFrame(x_test_sel,columns=columns)
```

```python
x_train_sel
```

```python
x_test_sel
```

```python
model = LinearRegression()

print("training",np.mean(cross_val_score(model,x_train_sel,y_train,cv=5,scoring='r2')))
print("testing",np.mean(cross_val_score(model,x_test_sel,y_test,cv=5,scoring='r2')))
```

# SEQUENTIAL BACKWARD SELECTION/ELEMINATION

```python
from sklearn.metrics import mean_squared_error
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
x=data.drop('medv',axis=1)
y=data['medv']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
x_train.shape
```

```python
sc = StandardScaler()
columns = x_train.columns
x_train = sc.fit_transform(x_train)
x_train = pd.DataFrame(x_train,columns=columns)
x_test = sc.transform(x_test)
x_test = pd.DataFrame(x_test,columns=columns)
x_train
```

```python
model = LinearRegression()
print("training",np.mean(cross_val_score(model,x_train,y_train,cv=5,scoring='r2')))
print("testing",np.mean(cross_val_score(model,x_test,y_test,cv=5,scoring='r2')))
```

```python
lr = LinearRegression()
sfs = SFS(lr,k_features='best',forward=False,floating=False,scoring='r2',cv=5)
model=sfs.fit(x_train,y_train)
```

```python
model.subsets_
```

```python
columns=model.k_feature_names_
```

```python
x_train_sel = model.transform(x_train)
x_train_sel = pd.DataFrame(x_train_sel,columns=columns)
x_train_sel.shape
```

```python
x_test_sel = model.transform(x_test)
x_test_sel = pd.DataFrame(x_test_sel,columns=columns)
```

```python
model = LinearRegression()
print("training",np.mean(cross_val_score(model,x_train_sel,y_train,cv=5,scoring='r2')))
print("testing",np.mean(cross_val_score(model,x_test_sel,y_test,cv=5,scoring='r2')))
```

#  SEQUENTIAL FORWARD SELECTION 

```python
x = data.drop('medv',axis=1)
y = data['medv']
```

```python
x_train,x1_test,y1_train,y1_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
sc=StandardScaler()
columns = x_train.columns
x_train = sc.fit_transform(x_train)
x_train = pd.DataFrame(x_train,columns=columns)
x_test = sc.transform(x_test)
x_test = pd.DataFrame(x_test,columns=columns)
print(x_train)
print(x_test)
```

```python
model = LinearRegression()
print("training",np.mean(cross_val_score(model,x_train,y_train,cv=5,scoring='r2')))
print("testing",np.mean(cross_val_score(model,x_test,y_test,cv=5,scoring='r2')))
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
sfs = SFS(model,k_features='best',forward=True,floating=False,scoring='r2',cv=5)
sel = sfs.fit(x_train,y_train)
```

```python
sel.subsets_
```

```python
columns = sel.k_feature_names_
print(columns)
```

```python
sel.transform(x_train)
x_train_sel = pd.DataFrame(x_train,columns=columns)
sel.transform(x_test)
x_test_sel = pd.DataFrame(x_test,columns=columns)
print(x_train_sel)
print(x_test_sel)
```

```python
model = LinearRegression()
print("training",np.mean(cross_val_score(model,x_train_sel,y_train,cv=5,scoring='r2')))
print("testing",np.mean(cross_val_score(model,x_test_sel,y_test,cv=5,scoring='r2')))
```
