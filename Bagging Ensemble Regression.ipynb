```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('bangluru_housing.csv')
```

```python
df
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor()
```

```python
lr.fit(x_train,y_train)
knn.fit(x_train,y_train)
dt.fit(x_train,y_train)
yp1=lr.predict(x_test)
yp2=knn.predict(x_test)
yp3=dt.predict(x_test)
print("r2_score of lr",r2_score(y_test,yp1))
print("r2_score of knn",r2_score(y_test,yp2))
print("r2_score of dt",r2_score(y_test,yp3))
```

```python
from sklearn.ensemble import BaggingRegressor

bag_reg = BaggingRegressor(random_state=1)
bag_reg.fit(x_train,y_train)
```

```python
ypb = bag_reg.predict(x_test)
print("r2_score of bagging reg",r2_score(y_test,ypb))
```
