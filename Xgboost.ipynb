# Xgboost Regressor

```python
import numpy as np
import pandas as pd
import time
```

```python
df = pd.read_csv('bangluru_housing.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import r2_score
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = XGBRegressor()
```

```python
x= df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
x
```

```python
y
```

```python
np.mean(cross_val_score(model,x,y,cv=10,scoring='r2'))
```

```python
model2 = RandomForestRegressor(max_depth=2)
```

```python
np.mean(cross_val_score(model2,x,y,cv=5,scoring='r2'))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model = XGBRegressor()
```

```python
start = time.time()
model.fit(x_train,y_train)
print("score = ",r2_score(model.predict(x_test),y_test))
print("time = ",time.time()-start)
```

```python
modelGB = GradientBoostingRegressor()
start = time.time()
modelGB.fit(x_train,y_train)
print("score = ",r2_score(modelGB.predict(x_test),y_test))
print("time = ",time.time()-start)
```

# Xgboost Classifier

```python
import numpy as np
import pandas as pd
import time
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
df = pd.read_csv('mnist_train.csv')
```

```python
d = df.sample(10000)
```

```python
x = d.iloc[:,1:]
y = d.iloc[:,0]
x.head()
y.head()
```

```python
start = time.time()
model = XGBClassifier()
np.mean(cross_val_score(model,x,y,cv=10,scoring='accuracy'))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = XGBClassifier()
```

```python
start = time.time()
model.fit(x_train,y_train)
print("accuracy = ",accuracy_score(model.predict(x_test),y_test))
print("time = ",time.time()-start)
```
