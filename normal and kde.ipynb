```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
from numpy.random import normal
```

```python
#creating a sample which is normally distributed with the help of normal 
#loc->mean scale->std
sample = normal(loc=50 , scale=5 , size=1000)
```

```python
sample
```

```python
sample_mean = sample.mean()
```

```python
sample_std = sample.std()
```

```python
#creating normal distribution with parameters as sample mean and smple std
from scipy.stats import norm
dist = norm(sample_mean,sample_std)
```

```python
values = np.linspace(sample.min(),sample.max(),100)
```

```python
probabilities = [dist.pdf(value) for value in values]
```

```python
probabilities
```

```python
plt.hist(sample,bins=10,density=True)
plt.plot(values,probabilities)
```

```python
sample1 = normal(loc=20 ,scale=5,size=300)
sample2 = normal(loc=40 ,scale=5,size=700)
csample = np.hstack((sample1,sample2))
```

```python
plt.hist(csample,bins=50)
```

```python
from sklearn.neighbors import KernelDensity
```

```python
model = KernelDensity(bandwidth=2 , kernel='gaussian')
```

```python
csample=csample.reshape(len(csample),1)
```

```python
model.fit(csample)
```

```python
values = np.linspace(csample.min(),csample.max(),100)
values = values.reshape(len(values),1)
```

```python
probabilities = model.score_samples(values)
probabilities = np.exp(probabilities)
```

```python
plt.hist(csample,bins=50,density=True)
plt.plot(values[:],probabilities)
plt.show()
```
